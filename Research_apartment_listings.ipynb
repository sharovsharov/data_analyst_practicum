```python
# check
# –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –±–∞—Ä—ã

def pass_value_barh(df):
    try:
        (
            (df.isna().mean()*100)
            .to_frame()
            .rename(columns = {0:'space'})
            .query('space > 0')
            .sort_values(by = 'space', ascending = True)
            .plot(kind= 'barh', figsize=(19,6), rot = -5, legend = False, fontsize = 16)
            .set_title('–ü—Ä–∏–º–µ—Ä' + "\n", fontsize = 22, color = 'SteelBlue')    
        );    
    except:
        print('–ø—Ä–æ–ø—É—Å–∫–æ–≤ –Ω–µ –æ—Å—Ç–∞–ª–æ—Å—å :) ')
```

```python
pass_value_barh(data)
```

```python
# –î–æ–±–∞–≤–∏–º —Å—Ç–æ–ª–±–µ—Ü —Å —Ü–µ–Ω–∞–º–∏ –æ–¥–Ω–æ–≥–æ –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ–≥–æ –º–µ—Ç—Ä–∞
data['price_per_m'] = data['last_price']/data['total_area']
# –î–æ–±–∞–≤–∏–º —Å—Ç–æ–ª–±–µ—Ü c –¥–Ω–µ–º –Ω–µ–¥–µ–ª–∏ –ø—É–±–ª–∏–∫–∞—Ü–∏–∏ –æ–±—ä—è–≤–ª–µ–Ω–∏—è (0 ‚Äî –ø–æ–Ω–µ–¥–µ–ª—å–Ω–∏–∫, 1 ‚Äî –≤—Ç–æ—Ä–Ω–∏–∫ –∏ —Ç–∞–∫ –¥–∞–ª–µ–µ);
data['weekday_exposition'] = data['first_day_exposition'].dt.weekday
# –ú–µ—Å—è—Ü–µ–º
data['month_exposition'] = data['first_day_exposition'].dt.month
# –ì–æ–¥–æ–º
data['year_exposition'] = data['first_day_exposition'].dt.year
```

```python
# –°–æ–∑–¥–∞–¥–∏–º —Ñ—É–Ω–∫—Ü–∏—é –∫–∞—Ç–µ–≥–æ—Ä–∏–∑–∞—Ü–∏–∏ –ø–æ —ç—Ç–∞–∂–∞–º
def floor_category(row):
    floors_total = row['floors_total']
    floor = row['floor']
    if floor == 1:
        return '–ø–µ—Ä–≤—ã–π'
    elif floor == floors_total:
        return '–ø–æ—Å–ª–µ–¥–Ω–∏–π'
    elif 1 < floor < floors_total:
        return '–¥—Ä—É–≥–æ–π'

# –î–æ–±–∞–≤–∏–º —Å—Ç–æ–ª–±–µ—Ü —Å –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º–∏ —ç—Ç–∞–∂–Ω–æ—Å—Ç–∏
data['floor_category'] = data.apply(floor_category, axis = 1)
```
